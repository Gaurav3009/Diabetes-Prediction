```python
_y = dataset['Outcome']
```

```python
dataset.drop('Outcome', axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m = MinMaxScaler()
```

```python
df1 = m.fit_transform(dataset)
```

```python
new_df = pd.DataFrame(df1, index = dataset.index, columns = dataset.columns)
```

```python
new_df.head()
```

```python
import seaborn as sns
def plot_graph(obj):
    sns.distplot(new_df[obj], label=obj)
```

```python
for i in new_df.columns:
    plot_graph(i)
```

```python
new_df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df, _y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
z = lr.predict(x_test)
```

```python
y = pd.DataFrame(z)
```

```python
y.rename(columns = {0 : 'Output'}, inplace = True)
```

```python
y.head(20)
```

```python
y['Output'].unique()
```

```python
y2 = pd.DataFrame(y_test)
```

```python
y2.head()
```

```python
a = np.arange(192)
```

```python
y2.set_index(a, inplace = True)
```

```python
y2.head(20)
```

```python
y = y.merge(y2, on = y.index)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, z))
```
